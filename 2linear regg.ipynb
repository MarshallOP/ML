```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('data/profit.csv')
population = df['population'].values.reshape(-1,1)
profit = df['profit'].values
```

```python
regressor = LinearRegression()
regressor.fit(population,profit)
```

```python
profit_predicted = regressor.predict(population)
```

```python
# model evaluation
print('Slope:' ,regressor.coef_)
print('Intercept:', regressor.intercept_)
print('Root mean squared error: ', mean_squared_error(profit, profit_predicted))
print('R2 score: ',r2_score(profit, profit_predicted))
```

```python
plt.scatter(population,profit)
plt.xlabel('population-Values from 0-25')
plt.ylabel('profit-values from 5-22.5')
 # predicted values
plt.plot(population,profit_predicted, color='r')
plt.show()
```
