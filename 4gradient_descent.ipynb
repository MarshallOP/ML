```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/profit.csv')
df.head()
```

```python
x= df['population']
y= df['profit']
```

```python
learningrate = 0.01
iteration = 1000
theta0 = 0
theta1 = 0
m = len(y)
cost_history = []
```

```python
for iteration in range(iteration):
    h = theta0 + theta1 * x
    cost = (1/(2*m))*np.sum((h-y)**2)
    cost_history.append(cost)
    theta0 = theta0 - learningrate*(1/m)*np.sum(h-y)
    theta1 = theta1 - learningrate*(1/m)*np.sum((h-y)*x)
    if iteration%100 == 0 :
        print(f'iteration: {iteration}, cost: {cost}')
        
```

```python
plt.scatter(x,y,label= 'training data')
plt.plot(x, theta0+theta1*x,color='red',label='linear regression')
plt.xlabel('population')
plt.ylabel('profit')
plt.legend()
plt.show()
```

```python
plt.plot(cost_history)
plt.xlabel('iteration')
plt.ylabel('Cost')
plt.show()
```
